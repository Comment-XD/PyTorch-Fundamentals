```python
web_cam = cv2.VideoCapture(0) 
# Links up to our webcam, for windows, webcam is considered video_device=0

while web_cam.isOpened(): # keeps looping when our webcam is opened
    ret, frame = web_cam.read()  # returns a return value and an image

    results = model(frame)
    
    cv2.imshow("YOLO", np.squeeze(results.render()) )
    # shows the image at that frame and "YOLO" is just the name of the pop up window
    model.conf
    # checks to see in a 10 seconds frame window whether or not the "q" is pressed
    if cv2.waitKey(10) & 0xFF == ord("q"): 
        break

web_cam.release()
cv2.destroyAllWindows()
        
```

```python
plt.imshow(np.squeeze(results.render()))
plt.axis("off")
plt.title("Web Cam")
plt.show()
object_type = results.names[int(np.squeeze(np.array(results.pred))[-1])]
confidence_value = np.squeeze(np.array(results.pred))[-2]
print("^ Boi look at this dude, only 88% person")
```
