```python
import torch 
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
model = torch.hub.load("ultralytics/yolov5", "yolov5s")
```

```python
img = "https://ultralytics.com/images/zidane.jpg"
```

```python
traffic_img = "https://www.transportation.gov/sites/dot.gov/files/pictures/shutterstock_168370127.jpg"
```

```python
results = model(img)
results.print()
```

```python
traffic_result = model(traffic_img)
traffic_result.print()
```

```python
traffic_result.xywh
```

```python
plt.imshow(np.squeeze(traffic_result.render()))
plt.axis("off")
```

```python
plt.imshow(np.squeeze(results.render()))
plt.show() 
```

```python
cap = cv2.VideoCapture(0) 
# Links up to our webcam, for windows, webcam is considered video_device=0

while cap.isOpened():
    ret, frame = cap.read() 

    cv2.imshow("YOLO", frame)

    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
        
```
